볼츠만 분포 pi는 시스템이 상태 i 에 있을 확률이고 εi는 해당 상태의 에너지이며 kT 는 볼츠만 상수 k 와 열역학적 온도 T 의 곱이다. 

image.png

```python
import matplotlib.pyplot as plt
import numpy as np

k = 1.38e-23

def Boltzmann(E, T):
    a = -E/(k*T)
    prob = np.exp(a)
    return prob

T = 4000
# T=4000K일 때, 4000K에 필요한 에너지일 확률~20000K에 필요한 에너지일 확률
Es = np.linspace(k*T, k*T*5, 100)
prob = Boltzmann(Es, T)

plt.plot(Es, prob, 'r-')
```
